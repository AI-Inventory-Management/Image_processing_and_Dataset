# Analysis of sodas dataset

Import data

```python
import numpy as np
import pandas as pd

df = pd.read_excel('./sodas_dataset_raw/session classes.xlsx')
df
```

```python
df.drop('Session number', axis=1, inplace=True)
df
```

```python
df = df[:12]
df
```

Identify unique values

```python
unique_values = []
unique_values += df['cell 1'].unique().tolist()
unique_values += df['cell 2'].unique().tolist()
unique_values += df['cell 3'].unique().tolist()
unique_values += df['cell 4'].unique().tolist()
unique_values += df['cell 5'].unique().tolist()
unique_values += df['cell 6'].unique().tolist()
unique_values += df['cell 7'].unique().tolist()
unique_values += df['cell 8'].unique().tolist()
unique_values = list(set(unique_values))
print(unique_values)
print(len(unique_values))
```

```python
def pretty_print_dict(dict):
    for key in dict:
        print("{k}:".format(k=key))
        print("\t{v}".format(v=dict[key]))

pretty_print_dict({"key1":1, "key2":2})
```

```python
sodas_dict = {}
for value in unique_values:
    for i in range(1,9):
        count = df[df['cell ' + str(i)] == value]['cell ' + str(i)].count()
        try:
            sodas_dict[value] += count
        except KeyError:
            sodas_dict[value] = count


pretty_print_dict(sodas_dict)
```
